```python
# some primes are truncatable in base 10: removing digits from the left or right results in other primes.
# left ex: 313, 13, 3
# right ex: 313, 31, 3
# there are only finitely many primes for which this is true on both sides. this program finds them.

import sympy as sp
import random as rand

def striptodigitsten(n):
    digits = []
    numberstub = n

    while numberstub > 0:
        d = numberstub % 10
        digits.insert(0, d)
        numberstub -= d
        numberstub /= 10
        numberstub = int(numberstub)

    return digits

def digitstoint(L):
    if type(L) != list:
        return -1
    if not all(type(d) == int for d in L):
        return -1

    length = len(L)
    total = 0
    for i in range(length):
        total += L[i] * 10 ** (length - 1 - i)

    return total


def isLtruncatableprime(n):
    if type(n) != int or n < 1 or not sp.isprime(n):
        return False

    digits = striptodigitsten(n)
    length = len(digits)

    if length == 1:
        return False

    while len(digits) > 0:
        #print(digits)
        number = digitstoint(digits)
        if not sp.isprime(number):
            return False
        digits = digits[1:]

    return True


def isRtruncatableprime(n):
    if type(n) != int or n < 1 or not sp.isprime(n):
        return False

    digits = striptodigitsten(n)
    length = len(digits)

    if length == 1:
        return False

    while len(digits) > 0:
        #print(digits)
        number = digitstoint(digits)
        if not sp.isprime(number):
            return False
        digits = digits[:-1]

    return True

#print(isLtruncatableprime(313), isRtruncatableprime(313))
primes = []
goodprimes = []
for i in range(2, 10000000):
    if sp.isprime(i):
        primes.append(i)

for p in primes:
    if rand.random() < 0.001:
        print('testing prime:', p)
    if isLtruncatableprime(p) and isRtruncatableprime(p):
        print('doubly truncatable prime:', p)
        goodprimes.append(p)

print(len(goodprimes), goodprimes, sum(goodprimes))
```
