```python
# calculates the maximal sum traversing the triangle that PE provided top to bottom.

import random
from google.colab import files

file = open('0067_triangle.txt')
trianglefile = list(file)
# print(len(trianglefile))
# 100
file.close()

# trianglelist = list(trianglefile[0].splitlines())
# print((trianglelist))
triangle = []

for i in range(100):
    triangle.append(trianglefile[i].split(' '))

# print(triangle[23])

for i in range(100):
    for j in range(i+1):
        if j < i:
            string = triangle[i][j]
            triangle[i][j] = int(string)  # need to turn these to ints
        if j == i:
            string = triangle[i][i]
            string = string[:-1]  # clip newline chars first this time
            triangle[j][j] = int(string)

# print(triangle[23])

triangleresults = []  # I can do this much faster if I cache results
                      # subsequent Lorxus sez: holy crap I wasn't expecting a *multiorder of magnitude speedup*, cool and good and nice and fun and genuinely extremely worth.

for i in range(100):
    triangleresults.append([-1] * (i + 1))

# print((triangleresults[23][2]))

triangleresults[0] = [triangle[0][0]]

def maxsumatloc(m, n):
    total = 0
    if m == 0 and n == 0:
        return triangle[0][0]  # base case
    elif triangleresults[m][n] > -1:
        return triangleresults[m][n]
    elif n == 0:
        for i in range(m):
            total += triangle[i][0]  # case: left edge
        triangleresults[m][0] = total + triangle[m][n]
        return total + triangle[m][n]
    elif m == n:
        for i in range(m):
            total += triangle[i][i]  # case: right edge
        triangleresults[m][m] = total + triangle[m][n]
        return total + triangle[m][n]
    else:
        # if random.random() < 0.0001:
        #     print('checking case:', m, n)
        triangleresults[m][n] = max(maxsumatloc(m-1, n-1), maxsumatloc(m-1, n)) + triangle[m][n]
        return max(maxsumatloc(m-1, n-1), maxsumatloc(m-1, n)) + triangle[m][n]  # whichever is larger, plus the cell of interest

# print(maxsumatloc(0, 0))
# print(maxsumatloc(2, 0))
# print(maxsumatloc(3, 3))
# print(maxsumatloc(1, 1))

bestbottomtot = -1
length = len(triangle)

for i in range(length):
    print('CHECKING COLUMN:', i)
    if maxsumatloc(length - 1, i) > bestbottomtot:
        bestbottomtot = maxsumatloc(length - 1, i)

print(bestbottomtot)
```
